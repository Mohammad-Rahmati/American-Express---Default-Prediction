```python
from evaluation_metric import *
from baseline_model import *
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
train_data = pd.read_pickle('Data/train_data_agg.pkl')
train_labels = pd.read_pickle('Data/train_labels.pkl').loc[train_data.index]

Features = train_data.columns
train_data.shape, train_labels.shape
```

```python
models, df_scores, importances, df_results = base_model_lgbm(train_data, train_labels, 20000)
```

```python
df_results
```

```python
test_data = pd.read_pickle('Data/test_data_agg.pkl')
```

```python
missing_cols = [f for f in Features if f not in test_data.columns]
test_data[missing_cols] = 0
```

```python
prediction_list = []
for keys in models.keys():
    prediction_list.append(models[keys].predict(test_data))
```

```python
prediction_df = pd.DataFrame(prediction_list).T
prediction_df.index = test_data.index
```

```python
prediction_df.max(axis=1).to_csv('Data/prediction_max.csv')
```
